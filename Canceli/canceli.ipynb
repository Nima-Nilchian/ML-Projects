<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
پیش‌پردازش و مهندسی ویژگی
</font>
</h2>


```python
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder, MinMaxScaler, StandardScaler

class Preprocessing():
    def __init__(self):
        self.selected_cols = ['Created', 'DepartureTime','Price', 'CouponDiscount', 'BillID', 'TicketID', 'ReserveStatus', \
                'Male',  'Domestic', \
                'VehicleClass', 'Vehicle', 'Cancel', 'TripReason']
        
        self.final_features = ['TimeDiff','net_price','ReserveStatus', 'has_discount', 'Male', 'Domestic', 'Cancel', 'TripReason', 
                               'is_with_familiy', 'Bus', 'Train', 'InternationalPlane', 'Plane']
        
        self.target_col = 'Cancel'
        
    def select_useful_cols (self, is_train) : 
        if is_train : 
            self.df = self.dataset[self.selected_cols]
        else : 
            test_useful_cols = self.selected_cols.copy()
            test_useful_cols.remove(self.target_col)
            self.df = self.dataset[test_useful_cols]
    
    def is_with_family(self):
        self.df['is_with_familiy'] = self.df.groupby('BillID').Male.transform('count') > 1
        
    def time_difference(self):
        self.df[['Created', 'DepartureTime']] = self.df[['Created', 'DepartureTime']].apply(pd.to_datetime)
        self.df['TimeDiff'] = (self.df['DepartureTime'] - self.df['Created']).dt.days
        self.df.drop(['Created', 'DepartureTime'], axis=1, inplace=True)
        
    def handle_monetary (self) : 
        self.df['net_price'] = self.df.Price - self.df.CouponDiscount
        self.df['net_price'] /= self.df['net_price'].max()
        self.df['has_discount'] = self.df.CouponDiscount > 0
    
    def encode_cities(self, is_train):
        if is_train:
            cities = list(set(self.df.From.append(self.df.To).values))
            self.city_encoder = LabelEncoder().fit(cities)
            
        city2idx = dict(zip(self.city_encoder.classes_, self.city_encoder.transform(self.city_encoder.classes_)))

        self.df['From_encoded'] = self.df['From'].map(city2idx).fillna(-1).astype(int)
        self.df['To_encoded'] = self.df['To'].map(city2idx).fillna(-1).astype(int)

        self.df.drop(['To', 'From'], axis=1, inplace=True)
    
    def normalize_cities(self, is_train):
        if is_train:
            cities_encoded = np.array(self.df.From_encoded.append(self.df.To_encoded).values).reshape(-1,1)
            self.city_scaler = StandardScaler().fit(cities_encoded)
            
        self.df['From_encoded'] = self.city_scaler.transform(self.df[['From_encoded']])
        self.df['To_encoded'] = self.city_scaler.transform(self.df[['To_encoded']])

        
    def encode_vehicle(self):
        vehic = pd.get_dummies(self.df.Vehicle)
        self.df = pd.concat([self.df, vehic], axis=1)
        self.df.drop(columns=['Vehicle'], inplace=True)
        
            
    def encode_categoricals(self, is_train):
        if is_train:
            self.cat_cols = self.df.select_dtypes(exclude=['int','float']).columns
            self.cat_les = {}

            for col in self.cat_cols : 
                le = LabelEncoder().fit(self.df[col])
                self.cat_les.update({col:le})
        
        for col in self.cat_cols :
            if col == self.target_col :
                print('col:', col)
                continue
            self.df[col] = self.cat_les[col].transform(self.df[col])
    
    def select_final_features (self, is_train) : 
        if is_train : 
            self.df = self.df[self.final_features]
        else : 
            test_final_cols = self.final_features.copy()
            test_final_cols.remove(self.target_col)
            self.df = self.df[test_final_cols]
            
    def transform(self, dataset:pd.DataFrame, is_train=True) : 
        self.dataset = dataset.copy()
        
        self.select_useful_cols(is_train)
        self.is_with_family()
        self.time_difference()
        self.handle_monetary()
        self.encode_vehicle()
        self.encode_categoricals(is_train)
        self.select_final_features(is_train)
        
        return self.df

        
```

```python
preprocessor = Preprocessing()
train_df = preprocessor.transform(train_data, is_train = True)
```

```python
train_df
```

```python
test_df = preprocessor.transform(test_data, is_train=False)
```

```python
# modeling
from sklearn.model_selection import train_test_split

trg = 'Cancel'

x_train, x_valid, y_train, y_valid = train_test_split(train_df.drop(trg, axis=1), train_df[trg] \
                                                      ,test_size=.2, stratify=train_df[trg])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

model = DecisionTreeClassifier(max_depth=12)
model.fit(x_train, y_train)
```

```python
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier

base_estimator = RandomForestClassifier(class_weight='balanced', n_estimators=50)
model2 = AdaBoostClassifier(base_estimator=base_estimator, n_estimators=2)
model2.fit(x_train, y_train)
```

<h3 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
معیار ارزیابی
</font>
</h3>


```python
# evaluate model
from sklearn.metrics import f1_score, classification_report, confusion_matrix

pred = model.predict(x_valid)
print(f1_score(y_valid, pred, pos_label=0))
print(classification_report(y_valid, pred))
confusion_matrix(y_valid, pred)
```

```python
from sklearn.metrics import f1_score, classification_report, confusion_matrix

pred = model2.predict(x_valid)
print(f1_score(y_valid, pred, pos_label=0))
print(classification_report(y_valid, pred))
confusion_matrix(y_valid, pred)
```
